```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("BootstrappedAndNoOutlier.csv", encoding = "utf-8",sep=",")
```

```python
data = data.drop('Unnamed: 0', axis = 1)
data
```

### Adding Caption_Length column 

```python
data['Caption_Length'] = data['Caption'].apply(len)
data
```

### Instagram engagement prediction model : Passive Aggressive Classifiers  

```python
#Importing relevant libraries 
from sklearn.linear_model import PassiveAggressiveRegressor
from sklearn.model_selection import train_test_split
```

Predicting Comments using Caption Length 

```python
x = np.array(data["Caption_Length"]) 
y = np.array(data["Comments"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=1000)
# reshape data to 1D array 
xtrain_reshaped = np.array(xtrain).reshape(-1, 1)
ytrain_reshaped = np.array(ytrain).reshape(-1, 1)
xtest_reshaped = np.array(xtest).reshape(-1, 1)
ytest_reshaped = np.array(ytest).reshape(-1, 1)
```

```python
model = PassiveAggressiveRegressor()
model.fit(xtrain_reshaped, ytrain_reshaped)
model.score(xtest_reshaped, ytest_reshaped)
```
